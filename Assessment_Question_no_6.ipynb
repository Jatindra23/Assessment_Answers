```python
file_path = '/content/drive/MyDrive/hi_IN.txt'

with open(file_path, 'r', encoding='UTF-8', errors='replace') as f:
    text = f.read()

import re
# Extract Hindi words using Unicode Devanagari range
words = re.findall(r'[\u0900-\u097F]+', text)
hindi_dict = set(words)
print(f'Number of unique Hindi words extracted: {len(hindi_dict)}')
```

```python
import pandas as pd

# Load your dataset file containing the 1,77,000 unique words (assumed single column CSV)
df_words = pd.read_excel('/content/drive/MyDrive/Unique Words Data.xlsx')  # replace with your file path

# Function to classify spelling
def classify_spelling(word):
    return 'correct spelling' if word in hindi_dict else 'incorrect spelling'

# Create a new column with classification
df_words['spelling_status'] = df_words['word'].apply(classify_spelling)

# Save the result as CSV with two columns: word, spelling_status
output_csv_path = '/content/drive/MyDrive/hindi_word_classification.csv'
df_words.to_csv(output_csv_path, index=False)

# Print summary
correct_count = (df_words['spelling_status'] == 'correct spelling').sum()
print(f'Number of correctly spelled unique words: {correct_count}')
print(f'Results saved to {output_csv_path}')
```
